# Putting it all together for mongo

```python
letters = 'abcdefghijklmnopqrstuvwxyz'
listlet = list(letters)
listlet.append('19')
```

```python
alpha_urls = ['http://www.azlyrics.com/'+let+'.html' for let in listlet]
```

```python
def find_artists(url):
    artist_links = []
    soup = BeautifulSoup(requests.get(url, headers=HEADERS).text, 'lxml')
    for col in soup.find_all('div', class_='col-sm-6 text-center artist-col'):
        for el in col.find_all('a'):
            artist = el.text
            art_link = el['href']
            artist_links.append((art_link, artist))
    return artist_links
```

```python
from time import sleep
from random import randint
import numpy as np

all_artists = []
seen_artists = set()
for alpha in alpha_urls:
    for artist in find_artists(alpha):
        if artist[0] not in seen_artists:
            seen_artists.add(artist[0])
            all_artists.append(artist)
            
```

```python
def find_songs(artist_tuple):
    song_dic = {}
    url = 'http://www.azlyrics.com/' + artist_tuple[0]
    soup = BeautifulSoup(requests.get(url, headers=HEADERS).text, 'lxml')
    if soup.find('title').text == u'ERROR: The requested URL could not be retrieved':
        return 'error'
    album_list = soup.find('div', id='listAlbum')
    if album_list:
        for el in album_list.find_all('a', id='', target='_blank'):
            sub_dic = {}
            song = el['href']
            sub_dic['artist'] = artist_tuple[1]
            sub_dic['title'] = el.text
            altype = re.search(r'(.+):', el.find_previous('div').text)
            year = re.search(r'\s+\((\d+)\)', el.find_previous('div').text)
            alname = re.search(r'\s+"(.+)"\s+', el.find_previous('div').text)
            if altype.group(1) == u'compilation':     # do not allow compilation since there would be repeats
                continue
            if altype:
                sub_dic['album_type'] = altype.group(1)
            if year:
                sub_dic['year'] = year.group(1)
            if alname:
                sub_dic['album_title'] = alname.group(1)
            song_dic[song] = sub_dic
    return song_dic
```
